```python
import pandas as pd

df = pd.read_csv("patients_datas.csv" , sep=";")

df = df.dropna(axis=1 , how="all")
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

df['heart_rate'] = pd.to_numeric(df['heart_rate'], errors='coerce')

df['temperature'] = df['temperature'].astype(str).str.replace(',', '.', regex=False)
df['temperature'] = pd.to_numeric(df['temperature'], errors='coerce')

df['age'] = pd.to_numeric(df['age'], errors='coerce')

df["HRCondition"] = "Normal"
df.loc[df["heart_rate"] < 60, "HRCondition"] = "Bradycardia"
df.loc[df["heart_rate"] > 100, "HRCondition"] = "Tachycardia"
df.loc[df["heart_rate"].isna(), "HRCondition"] = "Heart Rate Data Not Found"


print("After Heart Rate Condition:")
print(df)

df["TCondition"] = "Normal"
df.loc[df["temperature"] < 36, "TCondition"] = "Hypothermia"
df.loc[df["temperature"] > 38, "TCondition"] = "Hyperthermia"
df.loc[df["temperature"].isna(), "TCondition"] = "Temperature Data Not Found"


median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)



print("\nAfter Temperature Condition and Age Fillna:")
print(df)


risk_report = df.loc[df["TCondition"] == "Hyperthermia"].groupby("diagnosis").size().reset_index(name='count')

print("\nRisk Report for Hyperthermia by Diagnosis:")
print(risk_report)
```
